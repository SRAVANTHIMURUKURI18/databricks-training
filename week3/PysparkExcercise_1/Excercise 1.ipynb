##Using WITHCOLUMN

```python
# Create bonus column as 10% of salary
df.withColumn("bonus", col("salary") * 0.10).display()
```

```python
# Create annual_salary column
df.withColumn("annual_salary", col("salary") * 12).display()
```

```python
# Create tax column as 5% of salary
df.withColumn("tax", col("salary") * 0.05).display()
```

```python
# Create updated_salary by adding 5000
df.withColumn("updated_salary", col("salary") + 5000).display()
```

```python
from pyspark.sql.functions import when, col

# Create salary_category column
df.withColumn(
    "salary_category",
    when(col("salary") >= 80000, "High")
    .when(col("salary") >= 50000, "Medium")
    .otherwise("Low")
).display()
```

```python
# Create age_group column
df.withColumn(
    "age_group",
    when(col("age") < 25, "Young")
    .otherwise("Adult")
).display()
```

```python
# Create location column by combining city and department
from pyspark.sql.functions import col, concat_ws
df.withColumn(
    "location",
    concat_ws(" - ", col("city"), col("department"))
).display()
```

```python
# Create increment_salary column with 15% hike
df.withColumn(
    "increment_salary",
    col("salary") * 1.15
).display()
```

```python
from pyspark.sql.functions import col, when, year

# Create experience_status column based on joining year
df.withColumn(
    "experience_status",
    when(year(col("joining_date")) < 2021, "Experienced")
    .otherwise("Fresher")
).display()
```

```python
from pyspark.sql.functions import col, length

# Create name_length column using employee name
df.withColumn(
    "name_length",
    length(col("emp_name"))
).display()
```

```python
# Create is_high_salary column using condition
df.withColumn(
    "is_high_salary",
    when(col("salary") > 80000, True).otherwise(False)
).display()
```

```python
# Create salary_after_tax column
df.withColumn(
    "salary_after_tax",
    col("salary") - (col("salary") * 0.05)
).display()
```

```python
# Create department_code column
df.withColumn(
    "department_code",
    when(col("department") == "IT", "D001")
    .when(col("department") == "HR", "D002")
    .when(col("department") == "QA", "D003")
    .otherwise("D999")
).display()
```

```python
# Create double_salary column
df.withColumn(
    "double_salary",
    col("salary") * 2
).display()
```
